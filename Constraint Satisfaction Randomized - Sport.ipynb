```python
## read in from file

def read_csv(fileName):
    with open(fileName, 'r', encoding='utf-8-sig') as fileOb:
        text = fileOb.read()
        lines = text.split('\n')
        data = []; 
        for line in lines:
            data.append(line.split(','))
        return data
```

```python
## to write out to file

def write_csv(fileName, data):
    dataStr = ""
    for row in data:
        rowStr = ""
        for col in row:
            rowStr += col + ','
        else:
            rowStr = rowStr[:-1]
        dataStr += rowStr + '\n'
    else:
        dataStr = dataStr[:-1]
    with open(fileName, 'w') as fileOb:
        fileOb.write(dataStr)
```

```python
## backtracking_search method
def constraint_satisfaction_backtrack(game_assignment, weeks, game):
    
    ## recursively run algorithm until there are no more games
    if (game == len(game_assignment)):             
        return True
    
    ## to access constraints and fields
    global constraints
    global fields  
    
    ## game number in game_assignment
    game_num = constraints[game][0]
    
    ## the first team
    team1 = constraints[game][1]
    
    ## the second team
    team2 = constraints[game][2]
    
    ## posible weeks for game to be played
    possible_weeks = constraints[game][3:]               

    ## for each week in possible weeks that a game can be played
    for week in possible_weeks:
        
        ## if week is completely open, then assign game to that week
        if (weeks[week][0] == -1):  
            
            ## add week and field to game assignment
            game_assignment[game] = [game_num, team1, team2, week, fields[0]] 
            
            ## add field, team1, and team2 to weeks
            weeks[week][0] = [fields[0]]
            weeks[week][1] = [team1]
            weeks[week][2] = [team2]
            
            ## go to next game
            if (constraint_satisfaction_backtrack(game_assignment, weeks, game + 1)):
                return True                
            else:
                weeks[week][0] = -1
                game_assignment[game] = [game_num, team1, team2, -1, -1]
        
        ## if week already had games assigned
        elif (type(weeks[week][0]) == list):
            
            ## get the fields that are being played on this week and make a copy
            fields_playing_on = weeks[week][0]
            fields_playing_on_temp = fields_playing_on[:]
            
            ## if there are no more fields left, then continue
            if (len(fields_playing_on) == len(fields)):
                continue
                
            ## if a given team already has a game scheduled this week, then continue
            if (team1 in weeks[week][1] or team1 in weeks[week][2] or team2 in weeks[week][1] or team2 in weeks[week][2]):
                continue
            
            ## add team i and j to weeks
            weeks[week][1].append(team1)
            weeks[week][2].append(team2)
                
            ## add field that team i and j are playing on to fields in use that week
            fields_playing_on.append(fields[len(fields_playing_on)])
            
            ## add week and field to game assignment
            game_assignment[game] = [game_num, team1, team2, week, fields_playing_on[-1]]
            
            ## add updated fields_playing_on to weeks
            weeks[week][0] = fields_playing_on
            
            ## go to next game
            if (constraint_satisfaction_backtrack(game_assignment, weeks, game + 1)):
                return True
            else:
                weeks[week][0] = fields_playing_on_temp
                game_assignment[game] = [game_num, team1, team2, -1, -1]
        else:
            return False
```
